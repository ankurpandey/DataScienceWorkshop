```python
import pandas as pd
movies = pd.read_csv('movies.dat', delimiter='::',engine='python')
ratings = pd.read_csv('ratings.dat',delimiter='::',engine='python')
users = pd.read_csv('users.dat',delimiter='::',engine='python')
```

```python
merge_data = pd.merge(movies, pd.merge(ratings, users))
```
