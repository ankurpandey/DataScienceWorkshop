```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
kmeans_df = pd.read_csv('K_meansHW_TomMitchell_data1.csv',sep = ",",header=None,names=['x','y'])
```

```python
kmeans_df.head()
```

```python
%matplotlib inline
```

```python
import matplotlib as plt
```

```python
plt.pyplot.scatter(kmeans_df['x'],kmeans_df['y'])
```

```python
kmeans_result = KMeans(n_clusters = 2,random_state=0).fit(kmeans_df)
```

```python
kmeans_result.labels_
```

```python
kmeans_result
```

```python
plt.pyplot.scatter(kmeans_df['x'],kmeans_df['y'],c=kmeans_result.labels_)
```
